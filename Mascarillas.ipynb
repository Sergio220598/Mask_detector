## Model using mobilnetv2

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_dir="dataset/Train"
test_dir="dataset/Test"

train_datage=ImageDataGenerator(rescale=1/255)
test_datage=ImageDataGenerator(rescale=1/255, validation_split=0.2)

train_generator=train_datage.flow_from_directory(train_dir,
                                                 target_size=(224,224),
                                                 batch_size=16,
                                                 class_mode='categorical',
                                                 color_mode='rgb',
                                                 subset='training')
                                                
validation_generator=test_datage.flow_from_directory(test_dir,
                                                 target_size=(224,224),
                                                 batch_size=16,
                                                 class_mode='categorical',
                                                 color_mode='rgb',
                                                 subset='validation')

test_generator=test_datage.flow_from_directory(test_dir,
                                                 target_size=(224,224),
                                                batch_size=16,
                                                 class_mode='categorical',
                                                 color_mode='rgb')
```

```python
last_output=pre_trained_model.output
x=tf.keras.layers.Flatten()(last_output)
x=tf.keras.layers.Dense(32, activation='relu')(x)
x=tf.keras.layers.Dropout(0.2)(x)
x=tf.keras.layers.Dense(len(clases), activation='softmax')(x)

model_mobilnetv2=tf.keras.Model(pre_trained_model.input, x)
#model_mobilnetv2.summary()
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import imutils

def uso_del_modelo(image):
    proto = r"face_detector\deploy.prototxt"
    model = r"face_detector\res10_300x300_ssd_iter_140000.caffemodel"
    readNet=cv2.dnn.readNet(proto, model)
    blob = cv2.dnn.blobFromImage(image=image, scalefactor=1, size=(224, 224), mean=(104, 117, 123))
    readNet.setInput(blob)
    outputs=readNet.forward() #Aqui se obtiene las detecciones de cara
    return outputs
# outputs=uso_del_modelo(image)


def pos_caras(image,outputs):
    (h, w) = image.shape[:2] #alto y ancho de la imagen
    l_acierto=[]             #Lista los aciertos de caras (ejm-0.98)
    l_pos_caras=[]           #Posiciones de las caras
    l_caras_recortadas=[]    #Caras recortadas para introducir a la red neuronal de mascarillas

    #For dentreo de las 200 detecciones
    for i in range(outputs.shape[2]): 
        acierto=outputs[0,0,i,2]
        l_acierto.append(acierto)

        #Se obtiene las posiciones de las caras
        if acierto>0.5:          
            box = outputs[0, 0, i, 3:7] * np.array([w, h, w, h]) 
            (startX, startY, endX, endY) = box.astype("int")
            l_pos_caras.append((startX, startY, endX, endY))
    return l_pos_caras
# l_pos_caras=pos_caras(image,outputs)

# #Recorte de caras
def recorte_cara(image,l_pos_caras):
    (h, w) = image.shape[:2]
    l_caras_recortadas=[]
    for i in range(len(l_pos_caras)):
        (startX, startY, endX, endY)=l_pos_caras[i]
        
        #Asegurarse que los cuadrados esten dentro del marco
        (startX, startY) = (max(0, startX), max(0, startY))
        (endX, endY) = (min(w - 1, endX), min(h - 1, endY))
        
        #Recorte
        cara = image[startY:endY, startX:endX]
        cara = cv2.cvtColor(cara, cv2.COLOR_BGR2RGB)
        cara=cv2.resize(cara,(224,224)) #depende de la entra del modelo
        #cara = img_to_array(cara)
        cara = preprocess_input(cara) #de la libreria mobilnetv2
        l_caras_recortadas.append(cara)
    return l_caras_recortadas
# l_caras_recortadas=recorte_cara(l_pos_caras)

#Modelo
model_mobilnetv2 = tf.keras.models.load_model("modelo_mascarillas_v3.h5")

#Para los labels junto a los cuadrados
font = cv2.FONT_HERSHEY_SIMPLEX
fontScale = 0.8
color_si = (0, 255, 0)
color_no = (0, 0, 255)
thickness = 2

cap=cv2.VideoCapture(0)

while True:
    ret, frame=cap.read()
    #frame = imutils.resize(frame, width=800)
    output=uso_del_modelo(frame)
    posiciones=pos_caras(frame,output)
    cara_recortada=recorte_cara(frame,posiciones)
    
    #Asegurarse que detecto caras y uso del modelo de mascarillas
    if len(cara_recortada)>0:
        for i in range(len(cara_recortada)):
            cara=cara_recortada[i]
            cara = np.array(cara, dtype="float32")
            cara = np.reshape(cara,[1,224,224,3])
            preds = model_mobilnetv2.predict(cara, batch_size=32) #prediccion

            (startX, startY, endX, endY)=posiciones[i]
            pos = (startX, startY - 10)
            
            if preds[0][0]>0.5:
                #print("Si tiene mascarilla")
                cv2.rectangle(frame,(startX,startY),(endX,endY),color_si,2) 
                cv2.putText(frame, 'Con Mascarilla', pos, font, fontScale, color_si, thickness, cv2.LINE_AA)
            else:
                #print("No tiene mascarilla")
                cv2.rectangle(frame,(startX,startY),(endX,endY),color_no,2) 
                cv2.putText(frame, 'Sin Mascarilla', pos, font, fontScale, color_no, thickness, cv2.LINE_AA)

    cv2.imshow('camara',frame)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
